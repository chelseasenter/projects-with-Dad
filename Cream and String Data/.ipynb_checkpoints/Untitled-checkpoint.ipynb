```python
##Importing libraries

# Pandas is used for creating and manipulating DataFrames (aka Lists of lists, aka Rows + Columns) and Series (Lists)
import pandas as pd

#matplotlib will be used for data visualization (charts, graphs, etc)
import matplotlib.pyplot as plt

#scipy stats deals with statistical stuff
# i.e. mean median mode, zscore, standard error 
# it's newest library we've started using
import scipy.stats as sts

#numpy provides support for large, multi-dimensional arrays and matrices, 
#with mathmatical functions to operate these arrays
import numpy as np
```

```python
# create a variable that's equal to the file's pathway location (from where this ipynb file is located)
file_to_load = "Resources/cream-string-data.csv"

# Read file as csv
cm_stg_data = pd.read_csv(file_to_load)

#Create a dataframe to view the initial data
cm_stg_data_df = pd.DataFrame(cm_stg_data)
cm_stg_data_df
```

```python
cm_stg_data_df.describe()
cm_stg_data_df
```

```python
cm_stg_data_df['Month'] = pd.to_datetime(cm_stg_data_df['Date']).dt.month_name().str.slice(stop=3)
cm_stg_data_df
```

```python
cm_stg_data_df["Month Group"] = cm_stg_data_df.groupby(["Month"]).count()
cm_stg_df = cm_stg_data_df["Month Group"].count()

cm_stg_df
```
